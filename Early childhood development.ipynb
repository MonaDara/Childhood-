```python
df2 = df.sort_values(by=['Height of mother (cm)'])
plt.scatter(df2['Height of mother (cm)'], df2['The height of child at the end of 24 months old.'])
plt.xlabel('Mother height')
plt.ylabel('Child height 24 months old')
plt.grid()
plt.show()
```

```python
df3 = df.sort_values(by=['Height of father (cm)'])
plt.scatter(df3['Height of father (cm)'], df3['The height of child at the end of 24 months old.'])
plt.xlabel('Father height')
plt.ylabel('Child height 24 months old')
plt.grid()
plt.show()
```

```python
df4 = df.sort_values(by=['Baby’s Height at birth'])
plt.scatter(df4['Baby’s Height at birth'], df4['The height of child at the end of 24 months old.'])
plt.scatter(df4['Child’s gender'], df4['The height of child at the end of 24 months old.']color = 'g')
plt.xlabel('Baby’s Height at birth')
plt.ylabel('Child height 24 months old')
plt.grid()
plt.show()
```

```python
df5 = df.sort_values(by=['Age of mother in the time of conception?'])
plt.scatter(df5['Age of mother in the time of conception?'], df5['The height of child at the end of 24 months old.'])
plt.xlabel('Age of mother in conception')
plt.ylabel('Child height 24 months old')
plt.grid()
plt.show()
```

```python
df6 = df.sort_values(by=['Age of father in the time of conception?'])
plt.scatter(df6['Age of father in the time of conception?'], df6['The height of child at the end of 24 months old.'])
plt.xlabel('Age of father in conception')
plt.ylabel('Child height 24 months old')
plt.grid()
plt.show()
```
